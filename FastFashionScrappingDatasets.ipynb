## DATA SCRAPING FOR THE STUDY OF THE FAST FASHION UNIVERSE

```python
# Importation des bibliothèques
import requests
from bs4 import BeautifulSoup
import pandas as pd #pour le traitement des données et convertir nos fichier en csv
```

```python
# Envoyer une requête HTTP à la page web
url = 'https://zenodo.org/record/4261101#.ZDbPYXZBy3B'
response = requests.get(url)
html_text = response.text
```

```python
# Créer un objet BeautifulSoup à partir du contenu HTML de la page
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup) #Afficher soup
```

```python
# Extraction du titre de la page
title = soup.title.string
```

```python
# Afficher le titre de la page
print(title)
```

```python
#Avec l'id de la div de notre tableau on récupère le contenu de notre table que nous stockons dans une variable results
results = soup.find(id="collapseTwo")
```

```python
#Affichage des  resultas 
print(results) 
```

```python
#Nous avons stocké dans une variable(table)le contenu de la table 
table= results.find('table', {'class' : 'table table-striped'})
table 
```

```python
# Trouver les lignes de la table
rows = table.find_all('tr')
```

```python
#Afficher les cellules
rows
```

```python
#Nous remarquons que nos deux fichiers sont stockés dans une balise a . Il faut que nous recupérons les hrefs/liens pour récupérer nos fichiers.
for row in rows:
    link = row.find('a', {'class': 'filename'})
    if link:
        print(link.get('href'))
```

Pour récupérer les fichiers CSV de ces liens d'où se trouve nos deux fichiers, nous avons utilisé la bibliothèque requests de Python pour envoyer une requête GET au serveur en utilisant ces liens. Ensuite, nous avons utilisé la méthode content pour récupérer le contenu du fichier CSV sous forme binaire, ou utiliser la méthode text pour récupérer le contenu du fichier sous forme de chaîne de caractères. 

Avec nos conditions If on lui demande si le fichier est bien téléchargé d'affichier : Le fichier a été téléchargé avec succès sinon d'afficher : Erreur lors de la récupération du fichier. Comme nous pouvons le constacté nous avons téléchargé avec succès les deux fichiers qui se trouve maintenant dans notre dossier.

```python
url = 'https://zenodo.org/record/4261101/files/fastFashionCompDim.csv?download=1'
response = requests.get(url)

if response.status_code == 200:
    content = response.content
    # ou bien: content = response.text
    with open('fastFashionCompDim.csv', 'wb') as f:
        f.write(content)
    print('Le fichier a été téléchargé avec succès.')
else:
    print('Erreur lors de la récupération du fichier.')
```
